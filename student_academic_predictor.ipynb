```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
n = 10000
data = pd.DataFrame({
    "student_id": range(1, n+1),
    "study_hours": np.random.normal(4,1.5,n).clip(0,10),
    "attendance": np.random.normal(75,15,n).clip(40,100),
    "previous_score": np.random.normal(65,15,n).clip(30,100),
    "assignments_completed": np.random.randint(0,10,n),
    "sleep_hours": np.random.normal(7,1.2,n).clip(4,10),
    "internet_access": np.random.choice([0,1],n, p=[0.3,0.7]),
    "extracurricular": np.random.choice([0,1],n, p=[0.6,0.4]),
    "parental_support": np.random.choice([0,1],n, p=[0.4,0.6]),
})
```

```python
data
```

```python
score =(
    data["study_hours"] * 5 +
    data["attendance"] * 0.5 +
    data["previous_score"] * 0.5 +
    data["assignments_completed"] * 0.1 +
    data["parental_support"] * 0.2 +
    data["sleep_hours"] * 1 +
    data["internet_access"] * 0.1 +
    data["extracurricular"] * 0.1
)

data["final_score"] = np.where(score>75, "Pass", "Fail")

data.to_csv("student_data.csv", index=False)
```

```python
#missing values

for col in ["study_hours", "attendance", "previous_score"]:
  missing_idx =  np.random.choice(data.index, int(0.02 * n), replace=False)
  data.loc[missing_idx, col] = np.nan


#outliers
outlier_idx = np.random.choice(data.index, 50)
data.loc[outlier_idx, "study_hours"] = np.random.uniform(9,12,50)


#label noise (5% flip)
noise_idx = np.random.choice(data.index, int(0.05 *n), replace=False)
data.loc[noise_idx, "final_score"] = data.loc[noise_idx, "final_score"].map(
    {"Pass":"Fail", "Fail":"Pass"}
)

data.to_csv("updated_student_performance_noise.csv", index=False)
```

```python
data["study_hours"].fillna(data["study_hours"].median(), inplace=True)
data["attendance"].fillna(data["attendance"].median(), inplace=True)
data["previous_score"].fillna(data["previous_score"].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
Q1 = data["study_hours"].quantile(0.25)
Q3 = data["study_hours"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

data["study_hours"] = data["study_hours"].clip(lower, upper)
```
